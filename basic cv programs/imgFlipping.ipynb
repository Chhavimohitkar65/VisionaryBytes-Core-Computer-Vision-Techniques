```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Reading and displaying  the image
img = cv2.imread('photos\cartoon.jpeg')
name = 'cartoon'
plt.title('cartoon image')
plt.imshow(img)
```

```python
#filpping the image horizontally
img = cv2.imread('photos\cartoon.jpeg')
horizontal_flip = cv2.flip(img, 1)
name = 'flipping cartoon image horizontally'
plt.title(name)
plt.imshow(horizontal_flip)
```

```python
#flipping the image vertically
img = cv2.imread('photos\cartoon.jpeg')
flip_vertically = cv2.flip(img, 0)
name = 'fliping the cartoonn image vertically'
plt.title(name)
plt.imshow(flip_vertically)
```

```python
#flipping image vertically as well as horizontally 
img = cv2.imread('photos\cartoon.jpeg')
vertical_horizontal_flip = cv2.flip(img,-1)
name = ' flippling the image vertically as well as horizontally'
plt.title(name)
plt.imshow(vertical_horizontal_flip)
```
