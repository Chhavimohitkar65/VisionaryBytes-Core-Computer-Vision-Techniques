```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('cartoon.jpeg')
plt.title('original image')
plt.imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.title('rgb image')
plt.imshow(img_rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_gray = cv2.cvtColor(img_rgb , cv2.COLOR_RGB2GRAY)
plt.title('gray image')
plt.imshow(img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_blur = cv2.GaussianBlur(img_gray,(3,3),0)
plt.title('blur image')
plt.imshow(img_blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
edges = cv2.Canny(img_blur,threshold1=100,threshold2=200)
plt.title('edges image')
plt.imshow(edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
edges_rgb = cv2.cvtColor(edges,cv2.COLOR_BGR2RGB)
plt.title('edges in rgb image')
plt.imshow(edges_rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
