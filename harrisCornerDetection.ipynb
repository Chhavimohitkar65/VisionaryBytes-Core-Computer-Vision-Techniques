```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.imread('photos\house.jpg')
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#the input image should be of float so we use float 32 function from numpyto convert all grayscale values of img to float
img_gray = np.float32(img_gray)
plt.imshow(img_gray)
```

```python
#applying harris detection method
corner_img = cv2.cornerHarris(img_gray, 2,5,0.07)
#it is a fucntion whcih take input grayscale img as a parameter and block size as 2 and other alogo parameter
corner_img = cv2.dilate(corner_img, None)
#cv2.dilate apply morphological filter to the img
img[corner_img>0.01*corner_img.max()]=[0,0,255]
plt.imshow(img)
```

```python
rgb_img  = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```
